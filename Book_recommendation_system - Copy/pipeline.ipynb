```python

import pandas as pd
import numpy as np
import time
import os

# --- LIBRARIES CHECK ---
# We check for libraries to give you a clear error if they are missing
try:
    import faiss
    from sentence_transformers import SentenceTransformer
except ImportError:
    print("\n❌ MISSING LIBRARIES!")
    print("Please run this command in your terminal first:")
    print("pip install pandas sentence-transformers faiss-cpu")
    exit()

# --- CONFIGURATION ---
DATA_FILE = 'books_data.csv'       # The file you already have
INDEX_FILE = 'book_vectors.index'  # The file we will build

def main():
    # 1. LOAD DATA
    print(f"\n--- Step 1: Loading Data from {DATA_FILE} ---")
    if not os.path.exists(DATA_FILE):
        print(f"❌ Error: Could not find '{DATA_FILE}'.")
        print("   Make sure it is in the same folder as this script!")
        return
    
    df = pd.read_csv(DATA_FILE)
    print(f"✅ Loaded {len(df)} books.")

    # 2. VECTORIZE (THE BRAIN)
    print("\n--- Step 2: Vectorizing (This might take 2-3 minutes) ---")
    print("   Loading AI model 'all-MiniLM-L6-v2'...")
    model = SentenceTransformer('all-MiniLM-L6-v2')
    
    print("   Reading books and creating vectors...")
    start_time = time.time()
    # We encode the 'plot' column which contains Title + Author + Tags
    vectors = model.encode(df['plot'].tolist(), show_progress_bar=True)
    
    elapsed = time.time() - start_time
    print(f"✅ Vectorization complete. Created {len(vectors)} vectors.")
    print(f"   Time taken: {elapsed:.2f} seconds")

    # 3. BUILD INDEX (THE DB)
    print("\n--- Step 3: Building FAISS Index ---")
    dimension = vectors.shape[1]  # 384 for this model
    
    # Create the search index (Flat L2 = Exact Search)
    index = faiss.IndexFlatL2(dimension)
    
    # Add vectors to the index (Must be float32)
    index.add(np.array(vectors).astype('float32'))
    
    # Save to disk
    faiss.write_index(index, INDEX_FILE)
    print(f"✅ Success! Index saved to '{INDEX_FILE}'.")
    print("\n🎉 Sprint 1 Complete. You can now build the search app.")

if __name__ == "__main__":
    main()
```
